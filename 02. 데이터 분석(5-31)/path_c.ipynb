```python
path = []
for i in range(len(df_oxidation)):
    temp = ''
    temp = str(df_oxidation['Ox_Chamber'][i]) + str(df_photo_soft['photo_soft_Chamber'][i]) + \
    str(df_photo_litho['lithography_Chamber'][i]) + str(df_etching['Etching_Chamber'][i]) + \
    str(df_ion['Chamber_Num'][i])
    path.append(temp)
    
path
    
```

```python
path_group = df_inspect['Path'].value_counts()
path_group
```

```python
df_inspect.groupby('Path')['Target'].mean().min()
```

```python
df_inspect.groupby('Path')['Target'].mean().max()
```

```python
for path in path_group:
    df = df_inspect[df_inspect['Path'] == path]
    df['cnt'] = [i for i in range(1,len(df)+1)]
    
```

```python
# 결점수(불량수)의 모평균 (C)
target_mean = df_inspect['Target'].mean()
import math
#표준편차 루트C
target_std = math.sqrt(target_mean)

#관리 상한(UCL)
ucl = target_mean + 3*target_std
cl = target_mean
lcl = target_mean - 3*target_std
```

```python
dfn = df_inspect[df_inspect['Path'] == '21122']
dfn['cnt'] = [i for i in range(1,len(dfn)+1)]

plt.plot(dfn['cnt'], dfn['Target'],marker = 'o', markersize = 5)
plt.plot([0, len(dfn)],[ucl, ucl], "--g", linewidth=2)
plt.plot( [0, len(dfn)],[cl, cl], "--r", linewidth=4)
plt.plot( [0, len(dfn)],[lcl, lcl], "--g", linewidth=2)
plt.xlabel('num of wafer')
plt.ylabel('Target')
```
