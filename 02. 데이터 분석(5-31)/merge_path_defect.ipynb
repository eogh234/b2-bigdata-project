## path 리스트 만들기

```python
path = []
for i in range(len(df)):
    temp = ''
    temp = str(df['Ox_Chamber'][i]) + str(df['photo_soft_Chamber'][i]) + \
    str(df['lithography_Chamber'][i]) + str(df['Etching_Chamber'][i]) + \
    str(df['Chamber_Num'][i])
    path.append(temp)
    
path
```

```python
# 데이터 프레임에 추가
df['Path'] = path
df.head()
```
